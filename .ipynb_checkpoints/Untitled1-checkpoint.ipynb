```python
import pandas as pd

dados = {
    "mes": ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun"],
    "faturamento": [1822, 5654, 6428.59, 7100, 6800, 7500],
    "gastos": [1200, 2900, 3100, 3500, 3300, 3600]
}

df = pd.DataFrame(dados)
df
```

```python
df["lucro"] = df["faturamento"] - df["gastos"]
df["margem_%"] = (df["lucro"] / df["faturamento"]) * 100

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.lineplot(data=df, x="mes", y="faturamento", label="Faturamento", marker="o")
sns.lineplot(data=df, x="mes", y="gastos", label="Gastos", marker="o")
sns.lineplot(data=df, x="mes", y="lucro", label="Lucro", marker="o")

plt.title("Desempenho Financeiro Mensal")
plt.xlabel("MÃªs")
plt.ylabel("R$")
plt.grid(True)
plt.show()
```

```python
import ccxt
import pandas as pd

exchange = ccxt.binance()

ohlcv = exchange.fetch_ohlcv(
    symbol="BTC/USDT",
    timeframe="1d",
    limit=180  # Ãºltimos 180 dias
)

df = pd.DataFrame(
    ohlcv,
    columns=["timestamp", "open", "high", "low", "close", "volume"]
)

df["data"] = pd.to_datetime(df["timestamp"], unit="ms")
df.set_index("data", inplace=True)

df.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df.index, df["close"], label="BTC Close")
plt.title("Bitcoin â€“ PreÃ§o de Fechamento (USD)")
plt.xlabel("Data")
plt.ylabel("USDT")
plt.grid(True)
plt.legend()
plt.show()
```

```python
df["MM50"] = df["close"].rolling(50).mean()
df["MM200"] = df["close"].rolling(200).mean()

plt.figure(figsize=(12,6))
plt.plot(df.index, df["close"], label="PreÃ§o", alpha=0.6)
plt.plot(df.index, df["MM50"], label="MM50")
plt.plot(df.index, df["MM200"], label="MM200")

plt.title("BTC â€“ MÃ©dias MÃ³veis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_btc = df.copy()
df_btc = df_btc[["close"]].rename(columns={"close": "BTC"})
```

```python
import requests
import pandas as pd

url = "https://api.bcb.gov.br/dados/serie/bcdata.sgs.1/dados"
params = {
    "formato": "json",
    "dataInicial": df_btc.index.min().strftime("%d/%m/%Y"),
    "dataFinal": df_btc.index.max().strftime("%d/%m/%Y")
}

r = requests.get(url, params=params)
r.raise_for_status()

data = r.json()

df_usd = pd.DataFrame(data)
df_usd["data"] = pd.to_datetime(df_usd["data"], dayfirst=True)
df_usd["USD"] = df_usd["valor"].astype(float)

df_usd.set_index("data", inplace=True)
df_usd = df_usd[["USD"]]
```

```python
import numpy as np

datas = df_btc.index
cdi_anual = 0.10
cdi_diario = (1 + cdi_anual) ** (1/252) - 1

df_cdi = pd.DataFrame(index=datas)
df_cdi["CDI"] = (1 + cdi_diario) ** np.arange(len(datas))
```

```python
df_all = df_btc.join(df_usd, how="inner").join(df_cdi, how="inner")
df_all.head()
```

```python
df_norm = df_all / df_all.iloc[0] * 100
df_norm.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(df_norm.index, df_norm["BTC"], label="Bitcoin")
plt.plot(df_norm.index, df_norm["USD"], label="DÃ³lar")
plt.plot(df_norm.index, df_norm["CDI"], label="CDI")

plt.title("ComparaÃ§Ã£o de Performance â€“ BTC x USD x CDI (Base 100)")
plt.ylabel("Ãndice (Base 100)")
plt.xlabel("Data")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import requests
import pandas as pd

url = "https://api.bcb.gov.br/dados/serie/bcdata.sgs.12/dados"
params = {
    "formato": "json",
    "dataInicial": df_btc.index.min().strftime("%d/%m/%Y"),
    "dataFinal": df_btc.index.max().strftime("%d/%m/%Y")
}

r = requests.get(url, params=params)
r.raise_for_status()

data = r.json()

df_cdi = pd.DataFrame(data)
df_cdi["data"] = pd.to_datetime(df_cdi["data"], dayfirst=True)
df_cdi["cdi_pct"] = df_cdi["valor"].astype(float) / 100  # % â†’ decimal

df_cdi.set_index("data", inplace=True)
df_cdi = df_cdi[["cdi_pct"]]

df_cdi.head()
```

```python
df_cdi["CDI"] = (1 + df_cdi["cdi_pct"]).cumprod()
df_cdi = df_cdi[["CDI"]]

df_cdi.head()
```

```python
df_all = (
    df_btc
    .join(df_usd, how="inner")
    .join(df_cdi, how="inner")
)

df_all.head()
```

```python
df_norm = df_all / df_all.iloc[0] * 100
df_norm.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(df_norm.index, df_norm["BTC"], label="Bitcoin")
plt.plot(df_norm.index, df_norm["USD"], label="DÃ³lar")
plt.plot(df_norm.index, df_norm["CDI"], label="CDI (BCB diÃ¡rio)")

plt.title("BTC Ã— DÃ³lar Ã— CDI DiÃ¡rio Oficial (Base 100)")
plt.xlabel("Data")
plt.ylabel("Ãndice (Base 100)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
retorno = (df_all.iloc[-1] / df_all.iloc[0] - 1) * 100
retorno
```

```python
def max_drawdown(series):
    roll_max = series.cummax()
    drawdown = series / roll_max - 1
    return drawdown.min() * 100

dd = {
    "BTC": max_drawdown(df_all["BTC"]),
    "USD": max_drawdown(df_all["USD"]),
    "CDI": max_drawdown(df_all["CDI"]),
}

dd
```

```python
plt.figure(figsize=(12,6))

for col in ["BTC", "USD", "CDI"]:
    roll_max = df_all[col].cummax()
    drawdown = df_all[col] / roll_max - 1
    plt.plot(drawdown.index, drawdown * 100, label=col)

plt.title("Drawdown â€“ BTC x USD x CDI")
plt.ylabel("Drawdown (%)")
plt.xlabel("Data")
plt.grid(True)
plt.legend()
plt.show()
```

```python
aporte = 1000  # R$ 1.000 por mÃªs (valor nominal)
```

```python
def simular_dca(precos, aporte):
    cotas = 0
    historico = []

    for preco in precos:
        cotas += aporte / preco
        historico.append(cotas * preco)

    return historico
```

```python
df_dca = pd.DataFrame(index=df_m.index)

df_dca["BTC"] = simular_dca(df_m["BTC"], aporte)
df_dca["USD"] = simular_dca(df_m["USD"], aporte)
df_dca["CDI"] = simular_dca(df_m["CDI"], aporte)

df_dca.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(df_dca.index, df_dca["BTC"], label="Bitcoin (DCA)")
plt.plot(df_dca.index, df_dca["USD"], label="DÃ³lar (DCA)")
plt.plot(df_dca.index, df_dca["CDI"], label="CDI (DCA)")

plt.title("DCA Mensal â€“ BTC x USD x CDI")
plt.ylabel("PatrimÃ´nio acumulado (R$)")
plt.xlabel("Data")
plt.grid(True)
plt.legend()
plt.show()
```

```python
total_aportado = aporte * len(df_m)

resultado = pd.DataFrame({
    "Total Aportado (R$)": total_aportado,
    "Valor Final (R$)": df_dca.iloc[-1],
    "Retorno (%)": (df_dca.iloc[-1] / total_aportado - 1) * 100
})

resultado
```

```python
import requests
import pandas as pd

url = "https://api.coingecko.com/api/v3/coins/bitcoin/market_chart"
params = {
    "vs_currency": "usd",
    "days": 730,        # 2 anos
    "interval": "daily"
}

r = requests.get(url, params=params, timeout=30)
r.raise_for_status()
data = r.json()

df_btc = pd.DataFrame(data["prices"], columns=["timestamp", "BTC"])
df_btc["data"] = pd.to_datetime(df_btc["timestamp"], unit="ms")
df_btc.set_index("data", inplace=True)
df_btc = df_btc[["BTC"]]

df_btc.head()
```

```python
import requests
import pandas as pd
import time

# intervalo: Ãºltimos 2 anos
end = int(time.time())
start = end - 60 * 60 * 24 * 365 * 2  # 2 anos

url = "https://api.coingecko.com/api/v3/coins/bitcoin/market_chart/range"
params = {
    "vs_currency": "usd",
    "from": start,
    "to": end
}

headers = {
    "User-Agent": "Mozilla/5.0"  # MUITO IMPORTANTE
}

r = requests.get(url, params=params, headers=headers, timeout=30)
r.raise_for_status()

data = r.json()

df_btc = pd.DataFrame(data["prices"], columns=["timestamp", "BTC"])
df_btc["data"] = pd.to_datetime(df_btc["timestamp"], unit="ms")
df_btc.set_index("data", inplace=True)
df_btc = df_btc[["BTC"]]

df_btc.head()
```
