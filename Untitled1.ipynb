```python
import requests
import pandas as pd

url = "https://api.bcb.gov.br/dados/serie/bcdata.sgs.1/dados"
params = {
    "formato": "json",
    "dataInicial": df_btc.index.min().strftime("%d/%m/%Y"),
    "dataFinal": df_btc.index.max().strftime("%d/%m/%Y")
}

r = requests.get(url, params=params)
r.raise_for_status()

data = r.json()

df_usd = pd.DataFrame(data)
df_usd["data"] = pd.to_datetime(df_usd["data"], dayfirst=True)
df_usd["USD"] = df_usd["valor"].astype(float)

df_usd.set_index("data", inplace=True)
df_usd = df_usd[["USD"]]
```

```python
import numpy as np

datas = df_btc.index
cdi_anual = 0.10
cdi_diario = (1 + cdi_anual) ** (1/252) - 1

df_cdi = pd.DataFrame(index=datas)
df_cdi["CDI"] = (1 + cdi_diario) ** np.arange(len(datas))
```

```python
df_all = df_btc.join(df_usd, how="inner").join(df_cdi, how="inner")
df_all.head()
```

```python
df_norm = df_all / df_all.iloc[0] * 100
df_norm.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(df_norm.index, df_norm["BTC"], label="Bitcoin")
plt.plot(df_norm.index, df_norm["USD"], label="DÃ³lar")
plt.plot(df_norm.index, df_norm["CDI"], label="CDI")

plt.title("ComparaÃ§Ã£o de Performance â€“ BTC x USD x CDI (Base 100)")
plt.ylabel("Ãndice (Base 100)")
plt.xlabel("Data")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def max_drawdown(series):
    roll_max = series.cummax()
    drawdown = series / roll_max - 1
    return drawdown.min() * 100

dd = {
    "BTC": max_drawdown(df_all["BTC"]),
    "USD": max_drawdown(df_all["USD"]),
    "CDI": max_drawdown(df_all["CDI"]),
}

dd
```

```python
anos = 6

df_btc_usd = yf.download(
    "BTC-USD",
    period=f"{anos}y",
    interval="1d",
    auto_adjust=True,
    progress=False
)

# achata colunas se vier MultiIndex
if isinstance(df_btc_usd.columns, pd.MultiIndex):
    df_btc_usd.columns = df_btc_usd.columns.get_level_values(0)

df_btc_usd = df_btc_usd[["Close"]].rename(columns={"Close": "BTC_USD"})
df_btc_usd.index = pd.to_datetime(df_btc_usd.index)

df_btc_usd.head()
```

```python
url = "https://api.bcb.gov.br/dados/serie/bcdata.sgs.1/dados"
params = {
    "formato": "json",
    "dataInicial": df_btc_usd.index.min().strftime("%d/%m/%Y"),
    "dataFinal": df_btc_usd.index.max().strftime("%d/%m/%Y")
}

r = requests.get(url, params=params, timeout=30)
r.raise_for_status()

df_usdbrl = pd.DataFrame(r.json())
df_usdbrl["data"] = pd.to_datetime(df_usdbrl["data"], dayfirst=True)
df_usdbrl["USD_BRL"] = df_usdbrl["valor"].astype(float)
df_usdbrl = df_usdbrl.set_index("data")[["USD_BRL"]]

df_usdbrl.head()
```

```python
df_btc = df_btc_usd.join(df_usdbrl, how="inner")
df_btc["BTC_BRL"] = df_btc["BTC_USD"] * df_btc["USD_BRL"]
df_btc = df_btc[["BTC_BRL"]]

df_btc.head()
```

```python
url = "https://api.bcb.gov.br/dados/serie/bcdata.sgs.12/dados"
params = {
    "formato": "json",
    "dataInicial": df_btc.index.min().strftime("%d/%m/%Y"),
    "dataFinal": df_btc.index.max().strftime("%d/%m/%Y")
}

r = requests.get(url, params=params, timeout=30)
r.raise_for_status()

df_cdi = pd.DataFrame(r.json())
df_cdi["data"] = pd.to_datetime(df_cdi["data"], dayfirst=True)
df_cdi["cdi_pct"] = df_cdi["valor"].astype(float) / 100
df_cdi = df_cdi.set_index("data")[["cdi_pct"]]

df_cdi["CDI"] = (1 + df_cdi["cdi_pct"]).cumprod()
df_cdi = df_cdi[["CDI"]]

df_cdi.head()
```

```python
df_all = df_btc.join(df_cdi, how="inner")

df_norm = df_all.copy()
df_norm["BTC"] = df_norm["BTC_BRL"] / df_norm["BTC_BRL"].iloc[0] * 100
df_norm["CDI"] = df_norm["CDI"] / df_norm["CDI"].iloc[0] * 100

plt.figure(figsize=(12,6))
plt.plot(df_norm.index, df_norm["BTC"], label="Bitcoin (em BRL, Base 100)")
plt.plot(df_norm.index, df_norm["CDI"], label="CDI (Base 100)")
plt.title("BTC vs CDI â€“ Ãšltimos 6 anos (Base 100)")
plt.xlabel("Data")
plt.ylabel("Ãndice (Base 100)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
aporte = 500

def simular_dca(precos, aporte):
    cotas = 0.0
    valores = []
    for preco in precos:
        cotas += aporte / preco
        valores.append(cotas * preco)
    return valores

df_dca = pd.DataFrame(index=df_m.index)

# BTC: compra "cotas" de BTC (via preÃ§o BTC_BRL)
df_dca["BTC"] = simular_dca(df_m["BTC_BRL"], aporte)

# CDI: compra "cotas" do Ã­ndice CDI (ele jÃ¡ Ã© um Ã­ndice acumulado)
df_dca["CDI"] = simular_dca(df_m["CDI"], aporte)

df_dca.tail()
```
